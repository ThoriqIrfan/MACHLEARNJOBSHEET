```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = r"C:/Users/TUF/Documents/COOLEG/MACHLEARN/wbc.csv" # path dataset personalised for my PC, please change to your actual local directory
df = pd.read_csv(data) # load dataset

#taking away the ID number and anything outside the mean data

df = df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean' , 'smoothness_mean' , 'compactness_mean' , 'concavity_mean' , 'concave points_mean' , 'symmetry_mean' , 'fractal_dimension_mean']]

le = LabelEncoder() # initiate LabelEncoder object
df['diagnosis'] = le.fit_transform(df['diagnosis']) # encoding diagnosis

#Standardizing all columns except diagnosis

std = StandardScaler()
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['texture_mean'] = std.fit_transform(df[['texture_mean']])
df['perimeter_mean'] = std.fit_transform(df[['perimeter_mean']])
df['area_mean'] = std.fit_transform(df[['area_mean']])
df['smoothness_mean'] = std.fit_transform(df[['smoothness_mean']])
df['compactness_mean'] = std.fit_transform(df[['compactness_mean']])
df['concavity_mean'] = std.fit_transform(df[['concavity_mean']])
df['concave points_mean'] = std.fit_transform(df[['concave points_mean']])
df['symmetry_mean'] = std.fit_transform(df[['symmetry_mean']])
df['fractal_dimension_mean'] = std.fit_transform(df[['fractal_dimension_mean']])

df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

data = r"C:/Users/TUF/Documents/COOLEG/MACHLEARN/wbc.csv" # path dataset personalised for my PC, please change to your actual local directory
df = pd.read_csv(data) # load dataset

#taking away the ID number and anything outside the mean data

df = df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean' , 'smoothness_mean' , 'compactness_mean' , 'concavity_mean' , 'concave points_mean' , 'symmetry_mean' , 'fractal_dimension_mean']]

le = LabelEncoder() # initiate LabelEncoder object
df['diagnosis'] = le.fit_transform(df['diagnosis']) # encoding diagnosis

#Standardizing all columns except diagnosis

std = StandardScaler()
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['texture_mean'] = std.fit_transform(df[['texture_mean']])
df['perimeter_mean'] = std.fit_transform(df[['perimeter_mean']])
df['area_mean'] = std.fit_transform(df[['area_mean']])
df['smoothness_mean'] = std.fit_transform(df[['smoothness_mean']])
df['compactness_mean'] = std.fit_transform(df[['compactness_mean']])
df['concavity_mean'] = std.fit_transform(df[['concavity_mean']])
df['concave points_mean'] = std.fit_transform(df[['concave points_mean']])
df['symmetry_mean'] = std.fit_transform(df[['symmetry_mean']])
df['fractal_dimension_mean'] = std.fit_transform(df[['fractal_dimension_mean']])

#Stratify split of "diagnosis" with the split of 80:20

df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

#printing result of train-test split

print(f'Real Data Value:\n{df.diagnosis.value_counts()}')
print(f'Train Split of Data:\n{df_train.diagnosis.value_counts()}')
print(f'Test Split of Data:\n{df_test.diagnosis.value_counts()}')
```
